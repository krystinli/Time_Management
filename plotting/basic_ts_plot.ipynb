## 00_static_plot

```python
import pandas as pd
from datetime import date,timedelta
import matplotlib.pyplot as plt

data = pd.read_csv("../data/time_sheet.csv")
data.index = pd.to_datetime(data.Date) # set date as index
data.sort_index(inplace=True)
data.drop('Date', axis=1, inplace=True)
data.tail()
```

```python
def plot_static(data, colname, color, target):
    """
    plot column 
    save it under img/
    """
    fig, ax = plt.subplots()
    fig.set_size_inches(18, 5) # img size
    
    ax.plot(colname, 
        data = data, 
        color = "black", 
        linewidth = 1, 
        marker = 'o', 
        markeredgecolor = color,
        markersize = 3,)

    ax.set(xlabel = "Date", 
           ylabel = "Hours", 
           title = colname)
    
    plt.axhline(y=target, color='r', linestyle='dashed')
    plt.legend()

plot_static(data, "Total", "blue", 10)
```

## 01_bar_chart

```python
# leetcode monthly progress accumulated
x = ["2020-11", "2020-12", "2021-01", "2021-02", "2021-03",]
y = [18, 23, 0, 0, 0]

fig, ax = plt.subplots()
fig.set_size_inches(18, 5) # img size

plt.bar(x, y, 
        color="pink", 
        edgecolor="yellow",
        linewidth=2
       )

ax.set(xlabel = "Month", 
       ylabel = "Number of Questions", 
       title = "Leetcode Questions completed accumulative")

plt.show()
```

## 02_bokeh_plot

```python
import pandas as pd
from datetime import date

from bokeh.models.widgets import DateRangeSlider
from bokeh.layouts import layout
from bokeh.io import show, output_notebook, curdoc
output_notebook()
```

```python
data = pd.read_csv("../data/time_sheet.csv")
data.index = pd.to_datetime(data.Date)
data.sort_index(inplace=True)
data.drop('Date', axis=1, inplace=True)
data.tail()
```

```python
date_range_slider = DateRangeSlider(title="Date Range: ", 
                                    start=date(2020, 7, 6), 
                                    end=date.today(), 
                                    value=(date(2020, 7, 6), date(2020, 12, 26)), step=1)

l = layout(children=[[date_range_slider]], sizing_mode='fixed')
curdoc().add_root(l)
curdoc().title = "DateRangeSlider Example"

show(date_range_slider)
```
